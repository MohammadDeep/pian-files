```python
# prompt: plot y_norm[0][:, 2]  and [:,0] and figsize

import matplotlib.pyplot as plt
persion = 30
plt.figure(figsize=(20, 6))  # Adjust the figure size as needed
plt.plot(y_norm[persion][:, 2].cpu(), label='clear Heat')
#plt.plot(y_norm[persion][:, 0].cpu(), label='Heat ')
plt.plot(list_pain[persion].cpu(), label='clear pain')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of y_norm[0][:, 2] and y_norm[0][:, 0]')
plt.legend()
plt.grid(True)
plt.show()
```
